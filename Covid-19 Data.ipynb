```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

# NY Times Data

```python
NYT_data = pd.read_csv('https://raw.githubusercontent.com/nytimes/covid-19-data/master/rolling-averages/us.csv')
NYT_data.tail()
```

```python
NYT_data['deaths/cases_pcntg'] = (NYT_data['deaths_avg'] / NYT_data['cases_avg']) * 100
NYT_data.tail()
```

```python
plt.figure(figsize=(20,10))
plt.plot(NYT_data['date'], NYT_data['deaths_avg'], ls='--', label='AVG Deaths')
plt.xlabel('Date')
plt.ylabel('Rolling AVG Deaths')
plt.legend(loc='upper left')
plt.twinx()
plt.plot(NYT_data['date'], NYT_data['cases_avg'], label='AVG Cases', color='r')
plt.ylabel('Rolling AVG Cases')
plt.xticks(range(0, len(NYT_data['date']), 100))
plt.title('Rolling AVG Deaths and Cases')
plt.legend(loc='upper right')
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.plot(NYT_data['date'], NYT_data['deaths_avg'], ls='--', label='AVG Deaths')
plt.xlabel('Date')
plt.ylabel('Rolling AVG Deaths')
plt.legend(loc='upper left')

plt.twinx()
plt.plot(NYT_data['date'], NYT_data['deaths/cases_pcntg'], label='Prop Deaths', color='g')
plt.ylabel('Percent of Cases resulting in Death')
plt.legend(loc='upper right')
plt.title('Percent of Cases Resulting in Death')
plt.xticks(range(0, len(NYT_data['date']), 100))
plt.show()
```

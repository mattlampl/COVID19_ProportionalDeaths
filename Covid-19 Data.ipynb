```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

# NY Times Data

```python
NYT_data = pd.read_csv('https://raw.githubusercontent.com/nytimes/covid-19-data/master/rolling-averages/us.csv')
NYT_data.tail()
```

```python
NYT_data['deaths/cases_pcntg'] = (NYT_data['deaths_avg'] / NYT_data['cases_avg']) * 100
NYT_data.tail()
```

```python
plt.plot(NYT_data['date'], NYT_data['deaths_avg'], ls='--', label='Rolling AVG Deaths')
plt.xlabel('Date')
plt.ylabel('Rolling AVG Deaths')
plt.twinx()
plt.plot(NYT_data['date'], NYT_data['cases_avg'], label='Rolling AVG Cases', color='r')
plt.ylabel('Rolling AVG Cases')
plt.xticks([])
plt.title('Rolling AVG Deaths and Cases')
plt.show()
```

```python
plt.plot(NYT_data['date'], NYT_data['deaths_avg'], ls='--', label='Rolling AVG Deaths')
plt.xlabel('Date')
plt.ylabel('Rolling AVG Deaths')
plt.twinx()
plt.plot(NYT_data['date'], NYT_data['deaths/cases_pcntg'], label='Percentage Deaths per Case', color='g')
plt.ylabel('Percent of Cases resulting in Death')
plt.xticks([])
plt.title('Percent of Cases Resulting in Death')
plt.show()
```

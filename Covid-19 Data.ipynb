```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

# NY Times Data

```python
NYT_data = pd.read_csv('https://raw.githubusercontent.com/nytimes/covid-19-data/master/rolling-averages/us.csv')
NYT_data.tail()
```

```python
NYT_data['deaths/cases_pcntg'] = (NYT_data['deaths_avg'] / NYT_data['cases_avg']) * 100
NYT_data.tail()
```

```python
#fig = go.Figure()
fig = make_subplots(specs=[[{"secondary_y": True}]])
fig.add_trace(go.Scatter(x=NYT_data['date'], y=NYT_data['cases_avg'], name='New Cases',
                line= dict(color='#b53c3c', dash='dot', width=2)
                ), secondary_y=False)
fig.add_trace(go.Scatter(x=NYT_data['date'], y=NYT_data['deaths_avg'], name='New Deaths',
                line= dict(color='#65b0db', width=2)
                ), secondary_y=True)
fig.update_layout(
    title='Daily New Cases and Deaths',
    xaxis_title='Date',
)
fig.update_yaxes(title_text='Number of Cases', secondary_y=False)
fig.update_yaxes(title_text='Number of Deaths', secondary_y=True)

fig.show()
```

```python
fig = make_subplots(specs=[[{"secondary_y": True}]])
fig.add_trace(go.Scatter(x=NYT_data['date'], y=NYT_data['cases_avg'], name='New Cases',
                line= dict(color='#b53c3c', dash='dot', width=2)
                ), secondary_y=False)
fig.add_trace(go.Scatter(x=NYT_data['date'], y=NYT_data['deaths/cases_pcntg'], name='Proportion of Deaths to Cases',
                line= dict(color='#65b0db', width=2)
                ), secondary_y=True)
fig.update_layout(
    title='Proportion of COVID-19 Deaths to the Number of New Cases per Day',
    xaxis_title='Date',
)
fig.update_yaxes(title_text='Number of Cases', secondary_y=False)
fig.update_yaxes(title_text='Percent of Cases Resulting in Death', secondary_y=True)

fig.show()
```
